```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("CarsData.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
index = ["model", 'year', 'price', 'transmission', 'mileage', 'fuelType', 'tax',
       'mpg', 'engineSize', 'Manufacturer']

for i in index:
    print(df[i].value_counts(),'\n')
    print("-------------------------------------------------")
```

```python
sns.displot(df["fuelType"])
plt.show()
```

```python
def plot_catplot(X,k,he,a):
    #sns.set(font_scale=1.5)
    sns.catplot(data=df, x=X , kind=k, height=he, aspect=a)    
```

```python
plot_catplot("year","count",5,4)
```

```python
data = df.Manufacturer.value_counts()
plt.figure(figsize=(10.5,3))
x = list(data.index)
y = list(data.values)
x.reverse()
y.reverse()

plt.title("Most popular Brand")
plt.ylabel("Brands")
plt.xlabel("Number of cars sold")

plt.barh(x,y)
plt.show()
```

```python
def plot_violinplot(h,v,j):
    plt.figure(figsize=(15,8))
    sns.set(font_scale=1.5)
    sns.violinplot(data=df, x=h, y=v, hue=j)
    plt.title('Density and distribution of prices ', fontsize=15)
    plt.xlabel(h)
    plt.ylabel("Engine Size")
```

```python
plot_violinplot("Manufacturer","engineSize","transmission")
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='crest')
plt.show()
```
